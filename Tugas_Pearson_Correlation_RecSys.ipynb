```python
# Import those libraries
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
print("Libraries Imported!")
```

```python
# Import Dataset
data = pd.read_csv('https://raw.githubusercontent.com/X4D0/Rating-Prediction-PC/main/Dataset%20Pearson.csv?token=ALEMJIV7TUMDP55WDQOLYNLBOAED6')

# print dataframe.
data
```

```python
user = data.drop(columns=['NIM', 'Nama'])
user = user.loc[5]
neighbor = data.drop(columns=['NIM', 'Nama'])
neighbor = neighbor.drop(5)
print("User : \n",user)
print("\nNeighbor : \n",neighbor)
```

### Pearson Correlation

```python
# Convert dataframe into series
user_list = user.values.tolist()
neighbor_list = neighbor.values.tolist()

# Apply the pearsonr()
correlation = []
i = 1
for tetangga in neighbor_list:
  pearson, _ = pearsonr(user_list, tetangga)
  correlation.append(pearson)
  # print(tetangga,'\n')
  print('Tetangga',i,' Pearsons correlation: %.3f' % pearson)
  i+=1
  
```

```python
print('%.3f' % pearson)
```

```python
neighbor['Correlation'] = correlation
neighbor
```

### Neighborhood Selection

Sort by Correlation Values

```python
neighborhood = neighbor.sort_values('Correlation', ascending=False)
neighborhood
```

Select Top 5 (k=5)

```python
neighborhood_select = neighborhood[:5]
neighborhood_select
```

### Predicts the Ratings for Markisa from User 1301180034

```python
def ave_other_user(rating):
  return np.mean(rating)
  
average_user = np.mean(user)

total_corelasi_userNTetangga = 0
for tc in neighborhood_select['Correlation']:
  total_corelasi_userNTetangga+=tc

rumus_pembilang = 0 
for i in neighborhood_select.index: #by index selected neighbor
  rumus_pembilang = rumus_pembilang + (neighborhood_select.loc[i]['Correlation'] * (neighborhood_select.loc[i]['Markisa'] - ave_other_user(neighborhood_select.loc[i][0:10])))

print(rumus_pembilang)
print(total_corelasi_userNTetangga)

new_rating_pred = average_user + (rumus_pembilang/total_corelasi_userNTetangga)
print("New Rating Prediction: ",new_rating_pred)



```
